## **Задание 1.**
Условный оператор if-elif-else

```python
a = float(input('Оцените силу вашей душевной боли по шкале от 0 до 10: '))
if a < 0.7:
    print('Кажется, у вас всё очень даже неплохо! В честь этого купите себе чего-нибудь вкусного или красивого, вы этого заслуживаете! ')
elif 0.7 <= a <= 4.2:
    print('Разочарование или пассивное волнение? Что бы это ни было, может, вам сделать несколько глубоких вдохов и выдохов? Кто-то говорил, что это помогает...')
elif 4.2 < a <= 6.9:    
    print('Кажется, вы испытваете настоящий стресс. Посмотрите внутрь себя, может, пора найти причину и встретиться с ней один один? ')
elif 6.9 < a <= 8.0 :
  print('Вау, в игре настоящие эксперты по тревожности и апатии. Надеемся, у вас это временно... ')
elif 8.0 < a <= 10.0 :
  print('Сложно представить, как вы ещё держитесь! Но держитесь, не отпускайте руки, а лучше обратитесь за помощью :)')
else :
    print('Не хочется обесценивать ваши чувства, но мы вас не понимаем.')

```
